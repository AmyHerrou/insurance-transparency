```python
#Looking at total number of 'Plan_Number_Claims_Received' and 'Plan_Number_Claims_Denied'
#group by 'State', 'Issuer_Name', and 'Plan_Type'
plan= df.groupby('Plan_Type')['Plan_Number_Claims_Received', 'Plan_Number_Claims_Denied'].sum()
plan= plan.rename(columns = {'Plan_Number_Claims_Received':'Total_Plan_Number_Claims_Received', 'Plan_Number_Claims_Denied':'Total_Plan_Number_Claims_Denied'})
plan=pd.DataFrame(plan)
plan
```

```python
plan2= df.groupby(["State", "Issuer_Name", "Plan_Type"])['Plan_Number_Claims_Received', 'Plan_Number_Claims_Denied'].sum()
plan2= plan2.rename(columns = {'Plan_Number_Claims_Received':'Total_Plan_Number_Claims_Received', 'Plan_Number_Claims_Denied':'Total_Plan_Number_Claims_Denied'})
plan2=pd.DataFrame(plan2)
plan2
```

```python
plan2= plan2.sort_values(by=['Total_Plan_Number_Claims_Denied'], ascending=[False])
#plan2.head(10)
plan2
```
